```python
import pandas as pd
 
# Create a dictionary to hold the separate data frames
dfs = {}

# Loop through each region and create a separate data frame for it
for region in data['Region'].unique():
    region_df = data[data['Region'] == region].copy()
    region_df.reset_index(drop=True, inplace=True)
    dfs[region] = region_df

# Access the data frames for each region using their keys
northern_df = dfs['Northern']
western_df = dfs['Western']
southern_df = dfs['Southern']
eastern_df = dfs['Eastern']
northeastern_df = dfs['NorthEastern']
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data
df = pd.read_csv("C:/Users/PC/Desktop/Work/DVProject/Power/Power_Generation.csv")

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Calculate the total generation for each region by summing the columns
df['Total Generation Actual'] = df['Thermal Generation Actual (in MU)'] + df['Nuclear Generation Actual (in MU)'] + df['Hydro Generation Actual (in MU)']
df['Total Generation Estimated'] = df['Thermal Generation Estimated (in MU)'] + df['Nuclear Generation Estimated (in MU)'] + df['Hydro Generation Estimated (in MU)']

# Group the data by region and date, and calculate the mean of each group
grouped = df.groupby(['Region', 'Date']).mean().reset_index()

# Create a plot of the actual vs estimated generation for each region over time
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(15,10))
regions = ['Northern', 'Western', 'Southern', 'Eastern', 'NorthEastern']

for i, region in enumerate(regions):
    row = i // 3
    col = i % 3
    ax = axs[row, col]
    
    region_data = grouped[grouped['Region'] == region]
    ax.plot(region_data['Date'], region_data['Total Generation Actual'], label='Actual')
    ax.plot(region_data['Date'], region_data['Total Generation Estimated'], label='Estimated')
    ax.set_title(region)
    ax.set_xlabel('Date')
    ax.set_ylabel('Generation (MU)')
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from plotnine import *

# Read the data
df = pd.read_csv('C:/Users/PC/Desktop/Work/DVProject/Power/Installed_Power_Capacity_State_wise_MW.csv')

# Clean state names
df['state'] = df['state'].replace({
    'Andhra Pradesh.': 'Andhra Pradesh'
})

# Group by 'state' and calculate the sum of 'coal', 'nuclear', and 'renewable' for each state
df_grouped = df.groupby(['state']).sum()[['thermal_total', 'nuclear' ,'hydro','res']].reset_index()

# Calculate the percentage change in installed capacity for each state from the earliest year to the latest year
df_grouped['pct_change'] = (df_grouped['res'] - df_grouped['res'].iloc[0]) / df_grouped['res'].iloc[0] * 100

# Sort the states in descending order based on percentage change in installed capacity
df_top_five = df_grouped.nlargest(5, 'pct_change')

# Melt the dataframe to long format
df_top_five_melted = df_top_five.melt(id_vars='state', var_name='type', value_name='installed_capacity')
# Melt the dataframe to long format, selecting only 'state' and 'pct_change' columns
df_top_five_melted = pd.melt(df_top_five[['state', 'pct_change']], id_vars=['state'], var_name='type', value_name='pct_change')

# Create the bar plot
bar_plot = (ggplot(df_top_five_melted, aes(x='state', y='pct_change', fill='type'))
            + geom_bar(stat='identity', position='dodge', width=0.7)
            + labs(x='State', y='Percentage Increment of Renewable Resource (%)', title='Top Five Fastest-Growing Renewable States in India')
            + theme_minimal()
            + theme(axis_text_x=element_text(angle=90, hjust=1),
                    legend_position='right')
           )

print(bar_plot)

```

```python
# Group by 'state' and calculate the sum of each type of energy and the grand total
df_pie = df.groupby('region').mean()[['coal', 'gas', 'hydro', 'nuclear', 'res', 'grand_total']]

 

# Calculate the share of each type of energy out of the total capacity for each state
df_pie_pct = df_pie.apply(lambda x: x / x['grand_total'] * 100, axis=1)
df_pie_pct.drop('grand_total', axis=1, inplace=True)

# Define the size of the inner and outer circles
inner_radius = 0.4
outer_radius = 1.0

# Loop through each state and create a ring plot
for state in df_pie_pct.index:
    # Get the non-zero percentages for the state
    non_zero_pct = df_pie_pct.loc[state][df_pie_pct.loc[state] != 0.0]
    labels = non_zero_pct.index
    
    # Check if any part has a percentage of 0.0%
    if len(labels) == 0:
        continue
    
    # Create the outer circle
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.pie(non_zero_pct, labels=labels, radius=outer_radius, autopct='%1.1f%%', startangle=90, labeldistance=1.05, pctdistance=0.75)
    ax.set_title(f'Share of each type of energy in {state}')

    # Create the inner circle
    ax.pie([100], radius=inner_radius, colors='w')
    
    # Add a legend to the plot
    ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
    
    plt.show()
```
